```python
# import matplotlib library
import matplotlib.pyplot as plt

# allows for inline visualizations
# use plt.show() for other platforms
%matplotlib inline
```

```python
# import libraries
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Function method
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r')

plt.subplot(1,2,2)
plt.plot(y,x, 'b')
```

```python
# Object-oriented method
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
# blank figure
fig = plt.figure()

# adding axes
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) # add_axes(left, bottom, width, height)
axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_title('Larger Plot')
axes2.set_title('Smaller Plot')
```
